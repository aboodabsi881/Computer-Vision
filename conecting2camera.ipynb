```python
import cv2

cap = cv2.VideoCapture(0)

width  =int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # 1080
height =int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # 1920

# WINDOWS -- *'DIVX'
#MACOS OR LINUX -- *'XVID'

writer = cv2.VideoWriter('save_video.mp4',cv2.VideoWriter_fourcc(*'XVID'),20,(width,height)) #MORE FRAME PER SECOUND -> THE LARGER DATA GONNA BE

while True:

    ret,frame = cap.read()

    #operation (drawing)
    writer.write(frame)

    cv2.imshow('frame' , frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
writer.release()
cv2.destroyAllWindows()
```
