```python
import cv2
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
img1 = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Computer-Vision-with-Python/DATA/bricks.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
print("IMG1 :", img1.shape ,img1.dtype , ", IMG2" , img2.shape ,img2.dtype )
```

```python
img1 = cv2.resize(img1,(900,900))
img2 = cv2.resize(img2,(900,900))
print("IMG1 :", img1.shape ,img1.dtype , ", IMG2" , img2.shape ,img2.dtype )
```

```python
blended = cv2.addWeighted(img1,0.5,img2,0.5,0)
```

```python
plt.imshow(blended)
```
