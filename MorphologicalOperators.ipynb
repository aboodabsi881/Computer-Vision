```python
import cv2
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,"Test",(50,200),font,5,(255,255,255),15)
    return blank_img
```

```python
def display_img(img):
    plt.figure(figsize=(20,20))
    plt.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
result = cv2.erode(img,kernel,iterations=3)
```

```python
display_img(result)
```
