```python
img1 = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2 = cv2.resize(img2,(700,700))
```

```python
img1.shape
```

```python
x_offset = 0 
y_offset = 1300 - 700
```

```python
rows,cols,channels = img2.shape
```

```python
roi = img1[y_offset:1300,x_offset:700] #region of intrest
```

```python
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv =cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
mask_inv.shape
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape,255,dtype=np.uint8)
```

```python
white_background
```

```python
white_background.shape
```

```python
bk = cv2.bitwise_or(white_background,white_background,mask=mask_inv)
```

```python
plt.imshow(bk)
```

```python
fg  = cv2.bitwise_or(img2,img2,mask = mask_inv)
```

```python
plt.imshow(fg)
```

```python
final = cv2.bitwise_or(roi,fg)
```

```python
plt.imshow(final)
```

```python
larg_img = img1
small_img = final
```

```python
larg_img[y_offset:1300,x_offset:700] = small_img
```

```python
plt.imshow(larg_img)
```
