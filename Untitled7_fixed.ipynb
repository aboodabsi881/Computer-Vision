**Image preprocessing**

```python
# Keep only images if you want an image-only view of the dataset.
# The full pipeline below keeps question, answer, and image together.
images = ds.remove_columns(["question", "answer"])
```

```python
images
```

```python
from torchvision import transforms

image_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
def process_image(example):
    """Convert a PIL image to a normalized RGB tensor."""
    image = example["image"].convert("RGB")
    example["image"] = image_transform(image)
    return example
```

```python
# Apply image preprocessing to every split.
# Note: this can take a while for the full dataset.
image_dataset = ds.map(process_image)
```

**Text preprocessing**

```python
import string
import re
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

# Download required NLTK resources once.
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger")
nltk.download("averaged_perceptron_tagger_eng")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
print(set(string.punctuation))
```

```python
def get_pos(tag):
    """Map NLTK POS tags to WordNet POS tags for lemmatization."""
    if tag.startswith("J"):
        return wordnet.ADJ
    if tag.startswith("V"):
        return wordnet.VERB
    if tag.startswith("R"):
        return wordnet.ADV
    if tag.startswith("N"):
        return wordnet.NOUN
    return wordnet.NOUN


def clean_text(text):
    """Lowercase, remove punctuation/symbols, tokenize, POS-tag, and lemmatize."""
    text = str(text).lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()

    lemmatizer = WordNetLemmatizer()
    cleaned_sentences = []

    for sentence in sent_tokenize(text):
        tokens = word_tokenize(sentence)
        tags = pos_tag(tokens)
        lemmas = [lemmatizer.lemmatize(word, get_pos(tag)) for word, tag in tags]
        cleaned_sentences.append(" ".join(lemmas))

    return " ".join(cleaned_sentences)


def preprocessing_text(example):
    example["question"] = clean_text(example["question"])
    example["answer"] = clean_text(example["answer"])
    return example

# Use image_dataset so the processed data contains both transformed images and cleaned text.
processed_data = image_dataset.map(preprocessing_text)
```

```python
# Separate views of questions and answers, if needed.
questions = processed_data.remove_columns(["image", "answer"])
answer_text = processed_data.remove_columns(["image", "question"])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")


def tokenize(example):
    encoding = tokenizer(
        example["question"],
        padding="max_length",
        truncation=True,
        max_length=32,
    )

    example["input_ids"] = encoding["input_ids"]
    example["attention_mask"] = encoding["attention_mask"]
    return example

# Add BERT token IDs and attention masks.
dataset = processed_data.map(tokenize)
```

```python
# Build label mappings from the cleaned answers in all splits.
unique_answers = sorted(set(
    answer
    for split in processed_data.keys()
    for answer in processed_data[split]["answer"]
))

answer2id = {answer: idx for idx, answer in enumerate(unique_answers)}
id2answer = {idx: answer for answer, idx in answer2id.items()}


def encode_answer(example):
    example["label"] = answer2id[example["answer"]]
    return example

# IMPORTANT: map over `dataset`, not `processed_data`, so tokenized columns are preserved.
dataset = dataset.map(encode_answer)

dataset
```
