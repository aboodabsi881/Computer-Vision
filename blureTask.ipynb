```python
import cv2 
import numpy as np
import matplotlib.pylab as  plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg').astype(np.float32)/255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
def display_img(img):
    plt.figure(figsize=(15,15))
    plt.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
gamma = 2
```

```python
result = np.power(i,gamma)
```

```python
display_img(result)
```

```python
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,"Hello",(250,600),font,10,(255,0,0),5)
```

```python
display_img(img)
```

```python
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
det = cv2.filter2D(img,-1,kernel)
```

```python
display_img(det)
```
