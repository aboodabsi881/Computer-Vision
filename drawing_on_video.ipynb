```python
import cv2


#CALLBACK FUNCTION RECTANGLE 
def draw_rectamgle(evnet,x,y,flags,param):
    
    global pt1,pt2,topLeft_clicked,botRight_clicked

    if evnet == cv2.EVENT_LBUTTONDOWN:

        #RESET THE RECTUNGLE (IT CHECKS IF THE RECT THERE)
        if topLeft_clicked == True and  botRight_clicked == True:
            pt1 = (0,0)
            pt2 = (0,0)
            topLeft_clicked = False
            botRight_clicked = False
            
        
        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked =True

        elif botRight_clicked ==False:
            pt2 = (x,y)
            botRight_clicked =True
        
        


## GLOBAL VARIABLE
pt1 = (0,0)
pt2 = (0,0)
topLeft_clicked = False
botRight_clicked = False


## CONNECT TO THE CALLBACK
cap = cv2.VideoCapture(0)

# CRITICAL: The window name MUST match exactly in namedWindow, setMouseCallback, and imshow
cv2.namedWindow('Test') 
cv2.setMouseCallback('Test',draw_rectamgle)



while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # DRAWING ON THE FRAME BASED ON GLOBAL VARIABLES   
    # Draw a small red dot for the first click
    if topLeft_clicked:
        cv2.circle(frame, center=pt1, radius=5, color=(0, 0, 255), thickness=-1)

    # Draw the green rectangle once both clicks are registered
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, color=(0, 255, 0), thickness=3)

    cv2.imshow('Test', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
