```python
import cv2
import numpy as np

import matplotlib.pylab as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(200,200,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(100,400),radius=100,color=(75,80,200),thickness=8)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=4)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(400,100),radius=30,color=(100,80,200),thickness=-1)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
cv2.putText(blank_img,text = 'Hello',org=(10,200),fontFace=font,fontScale=2,color=(255,0,0),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
