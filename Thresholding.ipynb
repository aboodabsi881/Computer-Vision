```python
img = cv2.imread('Computer-Vision-with-Python/DATA/crossword.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_img(img):
    plt.figure(figsize=(14, 14))
    plt.imshow(img,cmap='gray')
```

```python
show_img(img)
```

```python
ret,thresh2 = cv2.threshold(img,190,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
show_img(thresh2)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
```

```python
show_img(th2)
```

```python
blended = cv2.addWeighted(thresh2,0.6,th2,0.4,0)
```

```python
show_img(blended)
```
