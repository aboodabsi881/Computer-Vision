```python
import numpy as np
import cv2
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def load_img(img_src):
    img = cv2.imread(img_src,0)
    #img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img = cv2.resize(img,(600,600))
    return img
```

```python
def display_img(img):
    plt.figure(figsize=(15,15))
    plt.imshow(img,cmap='gray')
    
```

```python
img = load_img("Computer-Vision-with-Python/DATA/sudoku.jpg")
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 5)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 5)
```

```python
display_img(sobely)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
blended = cv2.addWeighted(sobelx,0.3,sobely,0.6,0)
```

```python
display_img(blended)
```

```python
ret , th1 = cv2.threshold(blended,100,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernal = np.ones((5,5),np.uint8)
```

```python
grad = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernal)
```

```python
display_img(grad)
```
